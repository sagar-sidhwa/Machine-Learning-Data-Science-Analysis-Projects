## Importing the Required Libraries

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

## Loading the Dataset

```python
data = pd.read_csv('https://raw.githubusercontent.com/sagar-sidhwa/Machine-Learning-Data-Science-Analysis-Projects/refs/heads/main/delaney_solubility/delaney_solubility_data/delaney_solubility_with_descriptors.csv')
```

## EDA (Exploratory data Analysis)

```python
data.head()
```

```python
data.shape
```

```python
data_correlations = data.corr()['logS']
```

```python
data_correlations
```

## Splitting the DataSet

```python
y = data['logS']
```

```python
x = data.drop('logS', axis = 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state= 42)
```

## Training the Model

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

## Predicting the Results

```python
y_train_pred = lr.predict(x_train)
```

```python
y_test_pred = lr.predict(x_test)
```

## Evaluating Model Performance

```python
eval_model_performance = pd.DataFrame(columns = ['Model', 'mean_squared_error (Train)', 'r2_score (Train)', 'mean_squared_error (Test)', 'r2_score (Test)'])
```

```python
eval_model_performance.loc[len(eval_model_performance)] = ['Linear Regression', mean_squared_error(y_train, y_train_pred), r2_score(y_train, y_train_pred), mean_squared_error(y_test, y_test_pred), r2_score(y_test, y_test_pred)]
```

```python
eval_model_performance
```

## Visualization of Results

```python
plt.figure(figsize = (10,6))

plt.scatter(x = y_train, y = y_train_pred, alpha = 0.3)

z_train = np.polyfit(y_train, y_train_pred, 1)
p_train = np.poly1d(z_train)
plt.plot(y_train, p_train(y_train))

plt.xlabel("Actual logS Values (Train Data)")
plt.ylabel("Predicted logS Values (Train Data)")
plt.title("Train - Actual Results vs Predicted Results")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (10,6))

plt.scatter(x = y_test, y = y_test_pred, alpha = 0.3)

z_test = np.polyfit(y_test, y_test_pred, 1)
p_test = np.poly1d(z_test)
plt.plot(y_test, p_test(y_test))

plt.xlabel("Actual logS Values (Test Data)")
plt.ylabel("Predicted logS Values (Test Data)")
plt.title("Test - Actual Results vs Predicted Results")
plt.grid(True)
plt.show()
```
